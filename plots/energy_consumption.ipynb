```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 10, 1)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8, 5))

ax.plot(x, np.exp(x/1.8), lw=4, label="Chatbot user", c='orangered')

ax.plot(x, 7+ 5*np.log(x/1.8), lw=4, label="SupplySignal user", c='dodgerblue')

ax.set_xlim(1, 8.3)
ax.set_ylim(0, 100)
ax.set_axisbelow(True)
ax.grid(color='gray', linestyle='dashed', alpha=0.3)
ax.set_title('Energy consumption between individual users vs SupplySignal', size=14)
ax.set_xlabel('Number of news to be searched', size=14)
ax.set_ylabel('Relative energy consumption', size=14)
ax.set_yticks([])
ax.set_xticklabels([1] + [i * 10 for i in range(1, 8)])
ax.legend(prop={'size': 12})

fig.tight_layout()
plt.show()
fig.savef
```
